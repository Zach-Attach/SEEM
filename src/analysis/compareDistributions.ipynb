```python
import lib.distributions as distrib
import lib.degree as deg
import lib.distance as dist
import lib.graphs as gg
```

```python
%store -r SEEM_Array_Witvliet
%store -r SEEM_White

%store -r RDDAM_Graphs_W1
%store -r RDDAM_Graphs_W7
%store -r RDDAM_Graphs_W8
%store -r RDDAM_Graphs_White

%store -r rand_Graphs_W1
%store -r rand_Graphs_W7
%store -r rand_Graphs_W8
%store -r rand_Graphs_White

%store -r SEEM_Array_Witvliet_Rand
%store -r SEEM_White_Rand
```

```python
L1_Dict = {
  'CENN': gg.G_Witvliet[0],
  'SEEM': SEEM_Array_Witvliet[0],
  'RDDAM': RDDAM_Graphs_W1,
  'ERN': rand_Graphs_W1,
  'REEM': SEEM_Array_Witvliet_Rand[0]
}

L5_Dict = {
  'CENN': [gg.G_Witvliet[6], gg.G_Witvliet[7], gg.G_White],
  'SEEM': SEEM_Array_Witvliet[6] + SEEM_Array_Witvliet[7] + SEEM_White,
  'RDDAM': RDDAM_Graphs_W7 + RDDAM_Graphs_W8 + RDDAM_Graphs_White,
  'ERN': rand_Graphs_W7 + rand_Graphs_W8 + rand_Graphs_White,
  'REEM': SEEM_Array_Witvliet_Rand[6] + SEEM_Array_Witvliet_Rand[7] + SEEM_White_Rand
}
```

```python
def getSequences(data, measure):
  if type(measure) == list:
    return {
      'CENN': measure[0](data['CENN'],title='CENN', presentation=False, subplot=True),
      'SEEM': measure[1](data['SEEM'],title='SEEM', presentation=False, subplot=True),
      'RDDAM': measure[1](data['RDDAM'],title='RDDAM', presentation=False, subplot=True),
      'ERN': measure[1](data['ERN'],title='ERN', presentation=False, subplot=True),
      'REEM': measure[1](data['REEM'],title='REEM', presentation=False, subplot=True)
    }
  return {
    'CENN': measure(data['CENN'],title='CENN', presentation=False, subplot=True),
    'SEEM': measure(data['SEEM'],title='SEEM', presentation=False, subplot=True),
    'RDDAM': measure(data['RDDAM'],title='RDDAM', presentation=False, subplot=True),
    'ERN': measure(data['ERN'],title='ERN', presentation=False, subplot=True),
    'REEM': measure(data['REEM'],title='REEM', presentation=False, subplot=True)
  }

distSeqs_L1 = getSequences(L1_Dict, [dist.distanceSequence, dist.avgDistDist])
distSeqs_L5 = getSequences(L5_Dict, dist.avgDistDist)

degSeqs_L1 = getSequences(L1_Dict, [deg.degreeSequence, deg.avgDegDist])
degSeqs_L5 = getSequences(L5_Dict, deg.avgDegDist)
```

```python
distrib.wassersteinDistances(distSeqs_L1, 'Distance', 'L1')
distrib.wassersteinDistances(distSeqs_L5, 'Distance', 'L5')

distrib.wassersteinDistances(degSeqs_L1, 'Degree', 'L1')
distrib.wassersteinDistances(degSeqs_L5, 'Degree', 'L5')
```
